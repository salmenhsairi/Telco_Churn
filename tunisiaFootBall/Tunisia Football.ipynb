```python
#url link to the webpage : "Liste des matchs de l'equipe nationale tunisienne de football"
url = "https://fr.wikipedia.org/wiki/Liste_des_matchs_de_l%27équipe_de_Tunisie_de_football_par_adversaire"
```

```python
import requests
r = requests.get(url)
website = r.text

import pandas as pd
tables = pd.read_html( website, encoding="UTF-8")
```

```python
L = ['A','Algérie','Allemagne','Allemagne de l Est','Angleterre',
     'Angola','Arabie saoudite','Argentine','Australie','B','Belgique',
     'Brésil','C','Cameroun','Canada','Colombie','Costa Rica','E',
     'Égypte','Émirats arabes unis','Espagne','F','France','G','Gambie'
     ,'Ghana','I','Italie','J','Japon','L','Libye','M','Madagascar','Mali',
     'Maroc','Maurice','Mauritanie','N','Nigeria','Norvège','P','Panama',
     'Pays-Bas','Pologne','Portugal','R','République centrafricaine','Russie',
     'S','Sénégal','Serbie-et-Monténégro','Seychelles','Sierra Leone','Somalie',
     'Soudan','Suisse','U','Ukraine','Z','Zimbabwe']
```

```python
#Création d'un dictionnaire dont les clés sont des lettres et les valeurs
#sont les listes des pays par ordre alphabétique
dic = {}
for i in range(len(L)) :
        if len(L[i]) == 1:
            dic[L[i]] = []
            j=i+1
            while j != len(L) and len(L[j]) != 1:
                dic[L[i]].append(L[j])
                j+=1
```

```python
#Nettoyage des Tables
data=[]
for table in tables:
    if table.columns[0] == 'Date':
        data.append(table)
```

```python
#Création du dataFrame finale
dataSet = data[0]
for table in data[1:]:
    dataSet = dataSet.append(table)
```

## Création des filtres

```python
#Filtre sur les Matchs gagnées
def matchs_gagnées(df):
    subSet = df[ (df['Match'].str[-7:] == 'Tunisie') & (df['Score'].str[0] < df['Score'].str[-1])
           | (df['Match'].str[:7] == 'Tunisie') & (df['Score'].str[0] > df['Score'].str[-1])]
    return subSet
```

```python
matchs_gagnées(dataSet)
```

```python
#Filtre Sur Les Match perdus
def matchs_perdues(df):
    subSet = df[ (df['Match'].str[-7:] == 'Tunisie') & (df['Score'].str[0] > df['Score'].str[-1])
           | (df['Match'].str[:7] == 'Tunisie') & (df['Score'].str[0] < df['Score'].str[-1])]
    return subSet
```

```python
matchs_perdues(dataSet)
```

```python
#Filtre sur les Match Nulles 
```

```python
def matchs_nulles(df):
    subSet = df.drop(matchs_perdues(df).index,axis=0).drop(matchs_gagnées(df).index,axis=0)
    return subSet
```

```python
matchs_nulles(dataSet)
```

```python
#Extraction de Tout les nations
nat=[]
for l in dic.values():
    for i in range(len(l)):
        nat.append(l[i])
```

```python
#Filtre Par Equipe
def matchs_par_equipe(df,equipe):
    global nat
    if not equipe in nat:
        return False
    subSet = df[df['Match'].str.find(equipe) != -1]
    return subSet
```

```python
matchs_perdues(matchs_par_equipe(dataSet,'Sénégal'))
```

```python
#Filtre par Nombre de Buts 
def match_par_nbuts_tn(df,n):
    if not n in range(10):
        return False
    subSet = df[ (df['Match'].str[-7:] == 'Tunisie') & ( df['Score'].str[-1] == str(n))
           | (df['Match'].str[:7] == 'Tunisie') & (df['Score'].str[0] == str(n))]
    return subSet
```

```python
matchs_gagnées(match_par_nbuts_tn(dataSet,6))
```

```python
#Filtre de Date
def match_par_année(df,date):
    if not date in range(1960,2021):
        return False
    subSet = df[df.Date.str[-4:] == str(date) ]
    return subSet
```

```python
match_par_année(dataSet,2020)
```

```python
#Filtre Matchs interieurs
def matchs_interieurs(df):
    subSet = df[df.Lieu.str.find('Tunisie') != -1]
    return subSet
```

```python
matchs_interieurs(matchs_perdues(dataSet))
```

```python
#Filtre Matchs Exterieurs 
def matchs_exterieurs(df):
    subSet = df.drop(matchs_interieurs(df).index,axis=0)
    return subSet
```

```python
matchs_exterieurs(dataSet)
```
